```python
# skip_gram
import logging
from gensim.models import word2vec

def main():
    logging.basicConfig(format = "%(asctime)s : %(levelname)s : %(message)s", level = logging.INFO)
    sentences = word2vec.LineSentence("all-new.txt")
    model = word2vec.Word2Vec(sentences, vector_size=250, sg = 1)
    model.save("skip_gram_all-new.model")
    model.wv.save_word2vec_format("skip_gram_all-new_vector.txt")
=
if __name__ == "__main__":
    main()
```
